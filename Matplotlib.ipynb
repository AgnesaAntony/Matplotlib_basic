 # FORMAT STRINGS

fmt is the method used in format string,which means "marker|line|color"

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([3,6,8,9])
plt.plot(a,'o:r')
plt.show()
```

# Marker Size

```python
import matplotlib.pyplot as plt
import numpy as np

a=np.array([26,833,65])

plt.plot(a,marker="D",ms=15)
plt.show()
```

# Marker Edge colour

    The short form of Marker edge colour can be written as "mec"

```python
import matplotlib.pyplot as plt
import numpy

a= numpy.array([1,3,5,7])
plt.plot(a,marker="^",mec='m',ms=20)
plt.show()
```

# Marker face colour

    The shortform of marker face colour can be given by "mfc".

```python
import matplotlib.pyplot as plt
import numpy as np

a=np.array([2,5,7,9,0])
plt.plot(a,marker='>',mec='w',mfc='k',ms=20)
plt.show()
```

# Positioning the title

loc parameter are used for position .The  parameters are 'left','right','center'(Default).

```python
import matplotlib.pyplot as plt
import numpy as np
a= np.array([45,58,65,7])
b=np.array([89,79,67,25])

plt.plot(a,b)
plt.xlabel("points in x-axis")
plt.ylabel("points in y-axis")
plt.title("Position",loc="left")
plt.show()
```

# Grid

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([1,4,6,8,7])
b=np.array([5,9,3,5,8])

plt.xlabel("xpoint")
plt.ylabel("ypoint")
plt.title("Grid")
plt.grid()
plt.plot(a,b)
plt.show()
```

# Grid-> Axis

    We can use grid(axis='x') to plot the axis. The values are xaxis,yaxis,both(default)..

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([2,5,7,9,7])
b=np.array([6,8,9,4,8])

plt.xlabel("Xpoints")
plt.ylabel("ypoints")
plt.title("AxisGrid")
plt.plot(a,b)
plt.grid(axis='x')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([3,7,9,5])
b=np.array([4,6,8,7])

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("gird with color")
plt.plot(a,b)
plt.grid(axis='both',color='green',linestyle='-.',linewidth=0.5)
plt.show()
```

# SubPlot

# PieCharts

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([78,98,76])
plt.pie(a)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([12,45,75])
b=["Apple","Mango","Orange"]

plt.pie(a,labels=b)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([25,58,36,95])
b=["red","yellow","green","blue"]
plt.pie(a,labels=b,startangle=90)
plt.show()
```

```python
# explode function
import numpy as np
import matplotlib.pyplot as plt

a=np.array([20,57])
b=["car","bike"]
c=[0,0.2]

plt.pie(a,labels=b,explode=c)
plt.show()
```

```python
# LEGEND
import numpy as np
import matplotlib.pyplot as plt

a=np.array([85,25,64])
b=["CSE-A","CSE-B","CSE-C"]
c=["red",'blue','green']
d=[0.1,0,0]
plt.pie(a,labels=b,colors=c,explode=d)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

a=np.array([26,34,78])
b=["Apple","mango","grapes"]
c=["yellow","green","blue"]

plt.pie(a,labels=b,colors=c)
plt.legend(title="fruits:")
plt.show()
```
